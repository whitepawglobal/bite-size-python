## Load audio

```python
import librosa

audiopath = r"C:\Users\codenamewei\Documents\nlp-meeting-data\audio-chunks\timit\SX72.WAV.wav"

array, sampling_rate = librosa.load(audiopath)

print(type(array))
print(array.shape)
print(type(sampling_rate))
print(f"Sampling Rate: {sampling_rate}")
```

## Show Spectrogram

```python
import numpy as np
import matplotlib.pyplot as plt

# The spectrogram is created by taking short segments of the audio signal, typically lasting a few milliseconds, 
# and calculating the discrete Fourier transform of each segment to obtain its frequency spectrum. 
# The resulting spectra are then stacked together on the time axis to create the spectrogram
D = librosa.stft(array)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure().set_figwidth(12)
librosa.display.specshow(S_db, x_axis="time", y_axis="hz")
plt.colorbar()
```

## Show Log Mel Spectrogram

```python
S = librosa.feature.melspectrogram(y=array, sr=sampling_rate, n_mels=128, fmax=8000)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure().set_figwidth(12)
librosa.display.specshow(S_dB, x_axis="time", y_axis="mel", sr=sampling_rate, fmax=8000)
plt.colorbar()
```
