## Load audio

```python
import librosa

audiopath = r"C:\Users\codenamewei\Documents\nlp-meeting-data\audio-chunks\timit\SX72.WAV.wav"

array, sampling_rate = librosa.load(audiopath)

print(type(array))
print(array.shape)
print(type(sampling_rate))
print(f"Sampling Rate: {sampling_rate}")
```

## Show Waveform

```python
import matplotlib.pyplot as plt
import librosa.display

plt.figure().set_figwidth(12)
librosa.display.waveshow(array, sr=sampling_rate)
```

## Play soundfile

```python
from IPython.display import Audio

# Play the audio
Audio(array, rate=sampling_rate, autoplay=False)
```
